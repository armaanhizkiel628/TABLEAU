# q1 What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?

answer
1. ableau Public
* Anyone on the internet can view your dashboard.
* There is no access control—once published, its fully public.
* The underlying data is also accessible, meaning others can download or inspect it.
* Best for: portfolios, demos, non-sensitive data.

2. Tableau Server / Tableau Cloud
* Access is restricted to authorized users (via login, permissions, roles).
* You can control who sees what (user-level security, row-level security, etc.).
* Data is protected and not publicly exposed unless you choose to share it.
* Supports enterprise-grade security (authentication, encryption, governance).



# q2 When publishing a Tableau dashboard, what happens to a live connection compared to an extract? How does this affect data refresh and performance?

answer

1. Live Connection
* The dashboard stays connected to the original data source (database,warehouse, etc.).
* Every time a user opens or interacts with the dashboard, Tableau queries the live data.

2. Effects:
* Data freshness: Always up-to-date real-time or near real time.
* Performance: Can be slower, especially if:

   The database is large or complex

   Network latency is high
* Dependency: Requires constant access to the source system.

3. Extract Hyper Extract
* Tableau creates a snapshot of the data stored in its optimized format.
* The dashboard uses this stored copy instead of querying the live source.

4. Effects:
* Data freshness: Not automatic—you must schedule refreshes.
* Performance: Faster, because:

  Data is optimized for analytics

  Queries run locally within Tableau

* Flexibility: Works even if the original source is offline.

# q4 If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?

answer
1. You would adjust the “Download Data” permission in Tableau Server

2. What to do
* Allow: View (so users can see the dashboard)
* Deny or remove: Download Data

3. Tableau permissions are capability-based. Even if users can view and interact with a dashboard, removing Download Data ensures they cannot access or export the underlying dataset.

4. You may also want to review related permissions:

* Download Full Data → specifically controls access to full underlying data
* Download Summary Data → controls access to aggregated data

5. Keep View = Allowed, but Download Data = Denied to let users see the dashboard without exposing the data behind it.

# q6  List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.

answer

1. Remove or anonymize sensitive data
* Never include personally identifiable information (PII) like names, emails, phone numbers, or IDs.
* Aggregate or mask data (e.g., show totals instead of individual records).
* Double-check hidden fields—users may still access them if data is downloadable.



 Ensures no confidential or private data is exposed.

 2. Disable data download and limit exposure
* If using Tableau Server or Tableau Cloud, restrict permissions like:
   
   Download Data
   
   Download Workbook

* Hide unnecessary sheets, tooltips, or underlying data that could reveal more than intended.

* Prevents users from extracting or misusing the data.

# q7 After publishing a dashboard, users report a “Data not found” error.
*  What is the most common cause related to data source settings?




answer

1. The most common cause is a broken or incorrect data source connection after publishing.

2. What typically goes wrong


  When you publish to Tableau Server or Tableau Cloud, the data source settings may:

* Still point to a local file path (e.g., your computer)
* Use a database connection that the server cannot access
* Have missing or incorrect credentials


3. Why this causes “Data not found”

 The server tries to load the data but:

* The file isn’t available on the server
* The database isn’t reachable
* Authentication fails



* So Tableau can’t retrieve the data → resulting in the error.


4. How to fix it
* Ensure the data source is:

   Published along with the workbook as an extract, or

  Accessible from the server correct server/database path

* Embed or update valid credentials


* Verify connection settings in the published data source

   SSO Single Sign-On

 *  This verifies who the user is before granting access via Tableau Server or Tableau Cloud.


 3. 2. Permissions & Row-Level Security (Authorization)
