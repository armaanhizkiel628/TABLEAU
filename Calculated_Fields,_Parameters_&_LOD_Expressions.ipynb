# Q1 What is a Calculated Field and why is it important in business dashboards?


answer

1. A calculated field is a custom field you create in a dashboard or data tool like Excel using a formula based on existing data fields. Instead of being directly stored in the dataset, it is computed on the fly.


* important

2. Turn raw data into meaningful insights

* Raw numbers alone don’t tell much. Calculated fields help derive useful metrics like:

* Profit margins
* Growth rates
* Conversion rates

* This makes dashboards more insightful instead of just descriptive.

3. Enable real-time analysis

* Calculated fields update automatically when data changes.

 * If sales numbers change → profit updates instantly
No need to manually recalculate

4. Improve decision-making

* Business decisions rely on metrics, not raw data.Calculated fields provide:

* KPIs
* Ratios and comparisons
* Trends over time

* This helps managers quickly understand performance


5. A sales dashboard might include:

* Total Sales (raw data)
* Total Cost (raw data)
* Profit (calculated field)
* Profit Margin = (Profit / Sales) × 100 (calculated field)



# Q2 Why can’t parameters change visuals directly in Tableau?

ANSWER

1. A parameter is just a static input value, not part of the data model.

* It does not automatically interact with the data or charts
* It only changes something when you explicitly connect it

2. No direct link to the dataset

* Unlike filters or dimensions:

* Parameters are independent variables
* They are not tied to any column in your data

* So Tableau doesn’t know:

3. Filters and actions:

* Automatically update visuals because they operate on data

 Parameters:

* Do nothing by themselves
* They need a calculated field or parameter action to affect visuals

4. They don’t trigger queries alone

Tableau visuals update when:

* Data is filtered
* Fields change
* Queries are re-run

 Parameters:

* Don’t trigger these processes unless used in:
* Calculated fields
* Filters
* Reference lines



# q3 When should EXCLUDE LOD be used?

answer

1. To compare a value with an overall or higher-level value

Example:

* View: Sales by Region and Category

You want

* Sales of each category vs total sales per region

* EXCLUDE helps remove Category from the calculation so you get regional totals.

2. To calculate percentages of totals

Example:

* You want
* % contribution of each category to total sales in a region

* EXCLUDE lets you compute the denominator (total) by ignoring the category level

3. To create benchmarks or reference values

Example:

* Show each product’s sales
* Compare it to:

 Average sales across all products in that region

* EXCLUDE removes the product-level detail for the average.



# Q4 Difference between FIXED and INCLUDE LOD expressions?

answer

1. fixed lod

* What it does

Calculates values at a specific level, regardless of what’s in the view.

💡 Example

{ FIXED [Region] : SUM([Sales]) }

*  Always computes total sales per region, no matter:

* What dimensions are in the chart
* Whether Category or Product is shown

2. characteristic

* Ignores most dimensions in the view
* Respects context filters only
* Gives consistent, stable results

3. When to use FIXED
* when you need constant values
* When calculating:
* KPIs
* Totals
* Benchmarks
* When the calculation should not change with the view




4. Constant load
*  Adds extra granularity to the calculation, even if it’s not in the view.

 Example

* { INCLUDE [Customer Name] : SUM([Sales]) }

* Computes sales at customer level first, then aggregates back to the view level.


5. Key characteristics
* Depends on the dimensions in the view
* Adds finer detail → then rolls it up
* Affected by filters more naturall

6. When to use INCLUDE
* When you need:
* Averages of detailed data
* Per-customer or per-order calculations
* When the result should adapt to the view
